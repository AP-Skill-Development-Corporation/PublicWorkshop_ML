### 1. Standard Scaler
- Standard scaler will normally distribute our dataset.
- It scales all the data points between mean to unit variance
- **Formula**: (Xi - mean(X) ) / std(X)

```python
## Check dataset is normally distributed or not
plt.hist(df)
plt.show()
```

```python
df['TV'].mean()
```

```python
df['TV'].std()
```

```python
(230.1 - 147.0425) / 85.8542363149081
```

```python
### Apply standard scaler model
```

- step1: Import preprocessing model from sklearn
- step2: Initialize the model in any object
- step3: Fit the model 
- step4: Convert the array of values into dataframe

```python
from sklearn.preprocessing import StandardScaler
```

```python
std1 = StandardScaler()
```

```python
std2 = std1.fit_transform(df)
std2
```

```python
std2 = pd.DataFrame(std2, columns = df.columns)
std2
```

```python
import seaborn as sns
```

```python
fig, (ax1, ax2) = plt.subplots(ncols = 2, figsize = (9,5))
```

```python
fig, (ax1, ax2) = plt.subplots(ncols = 2, figsize = (9,5))

sns.kdeplot(df['TV'], ax = ax1)
sns.kdeplot(df['radio'], ax = ax1)
sns.kdeplot(df['newspaper'], ax = ax1)
sns.kdeplot(df['sales'], ax = ax1)

sns.kdeplot(std2['TV'], ax = ax2)
sns.kdeplot(std2['radio'], ax = ax2)
sns.kdeplot(std2['newspaper'], ax = ax2)
sns.kdeplot(std2['sales'], ax = ax2)
```

### MinMax scaler
- It scales all the data points between 0 to 1
- **Formul**: ( Xi - min(X) ) / ( max(X) - min(X) )

```python
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
ms1 = MinMaxScaler()
```

```python
ms2 = ms1.fit_transform(df)
ms2
```

```python
ms2 = pd.DataFrame(ms2, columns=df.columns)
ms2.head()
```
