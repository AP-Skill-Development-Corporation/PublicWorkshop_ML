# Categorical distribution plots:

 * boxplot() (with kind="box")

 * violinplot() (with kind="violin")

# Box Plot
- it mainly used for Statastical Problem
- it returns 3 quantiles 
    - 25%
    - 50% / median 
    - 75%
    - min
    - max
- it is used to find the Extreme values or Outliers
- same Matplotlib

```python
sns.catplot(y = "species", x = "petal_length", data = iris, kind = "box")
```

```python
sns.catplot(x = "species", y = "petal_length", data = iris, kind = "box", palette = ["y","r","g"])
```

```python
sns.catplot(data = iris, kind = "box", palette = ["y","r","g"])
```

```python
sns.boxplot(data= iris, palette = ["r","b","g","c"])
```

# Violin Plot
- categorial Distribution plot
- combination of both Box plot and Kernel Density Plot

```python
sns.catplot(x = "species", y= "sepal_length", data = iris, kind = "violin")
```

```python
sns.violinplot(x = iris.species, y = iris.sepal_length, palette = ["c","m","y"])
```

# Joint Plot
- Combination of kinds of plots
- it given much information
- it given Summary of dataset
- focus on single relationship

```python
sns.jointplot(x = "petal_length", y = "sepal_length", data = iris)
# default it returns scatter plot
```

```python
help(sns.jointplot)
```

```python
sns.jointplot(x = "petal_length", y = "sepal_length", data = iris, kind = "reg")
```

```python
sns.jointplot(x = "petal_length", y = "sepal_length", data = iris, kind = "kde")
```

```python
sns.jointplot(x = "petal_length", y = "sepal_length", data = iris, kind = "hex")
```

# Pair Plot
- combination 2 plots
- used to plot multiple plots at a time for the entire dataset
- scatter matrix
- it shows pairwise relationship for each variable/feaature

```python
sns.pairplot(data = iris, hue = "species",kind = "scatter",diag_kind="kde")
```

```python
sns.pairplot(data = iris, hue = "species",kind = "scatter",diag_kind="kde", markers = ["*","d","^"])
```

# Heatmap
- correlation Between the features
- relationship of two or more features / variables

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
tips.shape
```

```python
c = tips.corr()
# correlation can only handle numerical type data
c
```

```python
sns.heatmap(c)
# 1 means strongly correlated with each other
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (9,5))
sns.heatmap(c, annot = True)
```

```python
plt.figure(figsize = (9,5))
sns.heatmap(c, annot = True, cmap = "plasma", vmin = 0, vmax = 1)
```

```python
help(sns.heatmap)
```
