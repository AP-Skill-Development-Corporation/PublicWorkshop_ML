### Matplotlib:

* Visualizing of the data can be done using this Matplotlib library.
* This library is mainly used to visualize in the 2-d format.
* Pyplot module for visualization in matplotlib library.

**Different plotting methods**:
* Line plot
* Bar Graph
* Scatter plot
* Area plot
* Pie chart
* Histogram
* Box plot

**Line Plot**:
* Line plot is the linear grapg,in which all points are joined by a line
* plot() 

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]
plt.plot(x,y)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]
plt.plot(x,y,c = 'g')
plt.show()
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]
plt.plot(x,y,'g')
plt.show()
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]
plt.plot(x,y,color = '0')
plt.show()
```

```python
y = [i**2 for i in x]
plt.plot(x,y,c = 'y')
plt.title("Line plot")
plt.xlabel("x - values")
plt.ylabel("Power of x - values")
plt.show()
```

```python
import numpy as np
import pandas as pd
```

```python
x = np.linspace(0,4,30)
y = np.logspace(0,4,30)
print(x)
print(y)
```

```python
plt.plot(x,y,color = 'g',label = "Line plot -1")
plt.plot(x**2,y*3,color = 'm',label = "line plot -2")
plt.xlabel("X-axis")
plt.ylabel("y-axis")
plt.legend(loc = "best")
plt.show()
```

```python
plt.plot(x,y,color = 'g',label = "Line plot -1")
plt.plot(x**2,y*3,color = 'm',label = "line plot -2")
plt.xlabel("X-axis")
plt.ylabel("y-axis")
plt.legend(loc = "center")
plt.show()
```

```python
help(plt.plot)
```

```python
plt.plot(x,y,color = 'g',label = "Line plot -1",linestyle = "--")
plt.plot(x**2,y*3,color = 'm',label = "line plot -2",linestyle = "-.")
plt.xlabel("X-axis")
plt.ylabel("y-axis")
plt.legend(loc = "best")
plt.show()
```

```python
plt.plot(x,y,color = 'g',label = "Line plot -1",linestyle = "--",marker = "o")
plt.plot(x**2,y*3,color = 'm',label = "line plot -2",linestyle = "-.",marker = "s")
plt.xlabel("X-axis")
plt.ylabel("y-axis")
plt.legend(loc = "best")
plt.show()
```

```python
plt.plot(x,y,'#000000',label = "Line plot -1",linestyle = "--",marker = "o")
plt.plot(x**2,y*3,'#0000ff',label = "line plot -2",linestyle = "-.",marker = "s")
plt.xlabel("X-axis")
plt.ylabel("y-axis")
plt.legend(loc = "best")
plt.show()
```

**Subplots**:

* creating more plots with a plot we use subplots.
* subplots()

```python
fig,a = plt.subplots(2,2) 
a[0][0].plot(x,x*x)
a[0][0].set_title('Square')
a[0][1].plot(x,np.sqrt(x))
a[0][1].set_title('Square root')
a[1][0].plot(x,np.exp(x))
a[1][0].set_title('Exponential')
a[1][1].plot(x,np.log(x))
a[1][1].set_title("Logarithmic")
plt.show()
```

```python
fig
```

```python
a1 = plt.subplot2grid((3,3),(0,0),colspan =2)
a2 = plt.subplot2grid((3,3),(0,2),rowspan = 3)
a3 = plt.subplot2grid((3,3),(1,0),colspan = 2,rowspan = 2)
a2.plot(x,x*x,c = "g")
a2.set_title("Square")
a3.plot(x,np.exp(x),"m")
a3.set_title("Exponential")
a1.plot(x,np.log(x),c = "b")
a1.set_title("Logarithmic")
plt.show()
```

**Area Plot**:

stackplot()

```python
days = [1,2,3,4]
working = [8,11,9,10]
sleeping = [8,10,11,12]
plt.stackplot(days,working,sleeping,colors = ['g','m'],labels = ['Working',"sleeping"])
plt.legend()
plt.show()
```

**pie plot**:

pie()

```python
subjects = ['python','Java','C']
Marks = [90,50,30]
plt.pie(Marks,labels = subjects,colors = ['g','m','b'],radius = 1)
plt.show()
plt.savefig("piechart.png")
```

```python
subjects = ['python','Java','C']
Marks = [90,50,30]
plt.pie(Marks,labels = subjects,colors = ['g','m','b'],radius = 1,explode = [1,1,1])
plt.show()
```

```python
subjects = ['python','Java','C']
Marks = [90,50,30]
plt.pie(Marks,labels = subjects,colors = ['g','m','b'],radius = 1,explode = [0,0,1])
plt.show()
```

```python
subjects = ['python','Java','C']
Marks = [90,50,30]
plt.pie(Marks,labels = subjects,colors = ['g','m','b'],radius = 1,explode = [0,1,0])
plt.show()
plt.savefig("sample")
```

```python
subjects = ['python','Java','C']
Marks = [90,50,30]
plt.pie(Marks,labels = subjects,colors = ['g','m','b'],radius = 1,explode = [1,0,0])
plt.show()
```
